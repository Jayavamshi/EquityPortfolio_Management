# Equity Portfolio Management

```r
# Loading the required Libraries
library(dplyr)  
library(tidyverse)

IBM=read.csv("IBM.csv")
MSFT=read.csv("MSFT.csv")
GOOG=read.csv("GOOG.csv")
AAPL=read.csv("AAPL.csv")
AMZN=read.csv("AMZN.csv")
FB=read.csv("FB.csv")
NFLX=read.csv("NFLX.csv")
TSLA=read.csv("TSLA.csv")
ORCL=read.csv("ORCL.csv")
SAP=read.csv("SAP.csv")
YEN=read.csv("YEN.csv")

# Extracting Close and Adj.Close from all the stocks and Date from only IBM stock
IBM <- IBM[,c("Date","Close","Adj.Close")]
MSFT <- MSFT[,c("Close","Adj.Close")]
GOOG <- GOOG[, c("Close","Adj.Close")]
AAPL <- AAPL[, c("Close","Adj.Close")]
AMZN <- AMZN[, c("Close","Adj.Close")]
FB <- FB[, c("Close","Adj.Close")]
NFLX <- NFLX[, c("Close","Adj.Close")]
TSLA <- TSLA[, c("Close","Adj.Close")]
ORCL <- ORCL[, c("Close","Adj.Close")]
SAP <- SAP[, c("Close","Adj.Close")]

#Changing column names
colnames(IBM)=c('Date', 'IBM Close', 'IBM Adj.Close')
colnames(MSFT)=c( 'MSFT Close', 'MSFT Adj.Close')
colnames(GOOG)=c('GOOG Close', 'GOOG Adj.Close')
colnames(AAPL)=c('AAPL Close', 'AAPL Adj.Close')
colnames(AMZN)=c('AMZN Close', 'AMZN Adj.Close')
colnames(FB)=c('FB Close', 'FB Adj.Close')
colnames(NFLX)=c('NFLX Close', 'NFLX Adj.Close')
colnames(TSLA)=c('TSLA Close', 'TSLA Adj.Close')
colnames(ORCL)=c('ORCL Close', 'ORCL Adj.Close')
colnames(SAP)=c('SAP Close', 'SAP Adj.Close')

#Storing all the data in the newly created dataframe "TenStockData"
TenStockData <- data.frame(IBM, MSFT, GOOG, AAPL, AMZN, FB, NFLX, TSLA, ORCL, SAP); 
head(TenStockData)
```

```r
#InitialCash = 5million to start investing. Cash always stays in zero interest cash account
InitialCash=5000000

#TotalStockValue = Its the Stock Values we hold; (Number of shares*Price of share)
TotalStockValue=0

# Mark to Market Value 
MTM = InitialCash + TotalStockValue

#Rebalancing Days
DayIncrementor =5

#Dividing cash equally among the 5 shares
CashForEachShare=InitialCash/DayIncrementor;

#Initial set of shares
InitialCatalog=c("IBM","MSFT","GOOG","AAPL","AMZN")

#Storing all the purchased data in the newly created dataframe "PurchasedData"
PurchasedData<-data.frame(matrix(ncol = 4),stringsAsFactors = FALSE)
X <- c("PurchaseDate","Company","Volume","PurchasePrice")
colnames(PurchasedData) <- X
```

```r
#Calculating the difference percentage
PercentageChange<- function(a,b)
    {
    ((b-a)/a)*100
}
```

```r
#Creating Temporary Vectors
PurchaseDateTemp=c();
CompanyTemp=c();
VolumeTemp=c();
PurchasePriceTemp=c();

#Initilizer to start row in TenStockData
Present_Row=1

#Creating temporary vectors for percentage change 
MTMT=c();
MTMPCT=c();
```

```r
#Resets all vairables to default
reset<-function()
    {
    InitialCash<<-5000000
    TotalStockValue<<-0
    MTM <<- InitialCash + TotalStockValue
    DayIncrementor <<-5
    CashForEachShare<<-InitialCash/DayIncrementor;
    InitialCatalog<<-c("IBM","MSFT","GOOG","AAPL","AMZN")
    PurchasedData<<-PurchasedData[0,]
    PurchaseDateTemp<<-c();
    CompanyTemp<<-c();
    VolumeTemp<<-c();
    PurchasePriceTemp<<-c();
    Present_Row <<- 1
    
}
```

```r
#Updating the MTM,Cash, StockValue after every purchase
updateMTM<-function()
    {
    TotalShareValue=0
    for(i in 1:length(PurchasedData$PurchaseDate))
        {
        TotalShareValue=TotalShareValue+PurchasedData$Volume[i]*PurchasedData$PurchasePrice[i]
    }
    TotalStockValue<<-TotalShareValue
    InitialCash<<-InitialCash-TotalShareValue
    MTM<<-TotalStockValue+InitialCash
}
```

```r
#Outputs the number of share we can buy on specific date

Quantity <- function(DateToBuy,CashToBuy,ShareName,i){
    #Joining ShareName+".Close" so as ColumnName can refer in dataframe TenStockData 
    ColumnName=paste0(ShareName,".Close")
    
    ShareClosingPrice= TenStockData[TenStockData$Date == DateToBuy,ColumnName ]
    
    #floor value of the (cash/price)
    NumberOfShares=floor(CashToBuy/ShareClosingPrice)
    
    PurchaseDateTemp[i]<<-DateToBuy;
    CompanyTemp[i]<<-ShareName;
    VolumeTemp[i]<<-NumberOfShares;
    PurchasePriceTemp[i]<<-ShareClosingPrice;
    
    #Portfolio<<- rbind(Portfolio,c(DateToBuy,ShareName,NumberOfShares,ShareClosingPrice))
}
```

```r
#Updates the PurchasedData and adds all the new bought shares into PurchasedData
updatePurchasedData<-function()
    {
    PurchasedData<<-cbind.data.frame(PurchaseDate=PurchaseDateTemp,Company=CompanyTemp,Volume=VolumeTemp,PurchasePrice=PurchasePriceTemp,stringsAsFactors = FALSE)
    
}
```

```r
#Caluculate the percentage drop of Adj.Close Price
PerAdjCloseDrop<-function(D1,D2,column_name){  
    x1=TenStockData[TenStockData$Date == D1 ,c(column_name)]
    x2=TenStockData[TenStockData$Date == D2 ,c(column_name)]
    ((x2-x1)/x1)*100
}
```

```r
#Buying the shares
buy<-function(DateToBuy)
    {
    for(i in 1:length(InitialCatalog))
        {
            (Quantity(DateToBuy,CashForEachShare,InitialCatalog[i],i))
        }
    updatePurchasedData()
    updateMTM()
}
```

```r
#Updating The Initial Catalog according to buying_low and buying_high
UpdateInitialCatalog<-function(Present_Row,BuyStrategy,DayIncrementor)
    {
    Current_Date=TenStockData$Date[Present_Row]
    Last_Date=TenStockData$Date[Present_Row-DayIncrementor]
    share = c("IBM", "MSFT", "GOOG", "AAPL", "AMZN", "FB", "NFLX", "TSLA", "ORCL", "SAP");
    percent = c(PerAdjCloseDrop(Last_Date,Current_Date, "IBM.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "MSFT.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "GOOG.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "AAPL.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "AMZN.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "FB.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "NFLX.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "TSLA.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "ORCL.Adj.Close"),PerAdjCloseDrop(Last_Date,Current_Date, "SAP.Adj.Close"))
    DF <- data.frame(share, percent,stringsAsFactors = FALSE) 
    if(BuyStrategy=="buying_low")
        {
        DF <- DF[order(DF$percent),]
    }
    else
        {
        DF <- DF[order(DF$percent,decreasing = TRUE),]
    }
    InitialCatalog<<-DF$share[1:5]
}
```

```r
#Sell all the shares in the PurchasedData and update the MTM, InitialCash, TotalStockValue
SellAllShares<-function(Present_Row)
    {
    Current_Date=TenStockData$Date[Present_Row]
    
    TotalShareValue=0
    for(i in 1:length(PurchasedData$PurchaseDate))
        {
        curr_column = paste0(PurchasedData$Company[i], ".Close")
        Current_Price= TenStockData[TenStockData$Date==Current_Date, curr_column]
        TotalShareValue=TotalShareValue+PurchasedData$Volume[i]*Current_Price
    }
    InitialCash<<-InitialCash+TotalShareValue
    TotalStockValue<<-0
    MTM<<-TotalStockValue+InitialCash
    
}
```

```r
#Calculate the Dividend
DividendCalculator= function(name,i)
    {
    close = paste0(name, ".Close")
    AdjClose = paste0(name, ".Adj.Close")
    
    (((TenStockData[i-1,close]/TenStockData[i,close]) - (TenStockData[i-1,AdjClose]/TenStockData[i,AdjClose]))* TenStockData[i,close])

}
```

```r
#Update Dividend after every Dividend day and add the Dividend to InitialCash
UpdateDividend<-function(Present_Row)
    {
    for(i in (Present_Row-(DayIncrementor-1)):Present_Row)
        {
        for(j in 1:length(PurchasedData$PurchaseDate))
            {
            name=PurchasedData$Company[j]
            Dividend= DividendCalculator(name,i )
            if(Dividend>0.1)
                {
                DividendAmount=Dividend* PurchasedData$Volume[j]
                InitialCash<<-InitialCash+DividendAmount
            }
        }
    }
}
```

```r
#Finding Daily MTM
DailyMTMCalculator<- function(index,DayIncrementor)
    {
    
    TotalStockValueT=c((1:(DayIncrementor-1)*0))
    for(i in 1:length(PurchasedData$PurchaseDate))
        {
        
        col_name= paste0(PurchasedData$Company[i], ".Close")
        
        for(j in 1:(DayIncrementor-1))
            {
            TotalStockValueT[j]= TotalStockValueT[j] + TenStockData[index-DayIncrementor+j,col_name]*PurchasedData$Volume[i]
        }
        
    }
    for(j in 1:(DayIncrementor-1))
        {
        MTMT[index-DayIncrementor+j]<<-TotalStockValueT[j]+InitialCash
    }
    
    for(j in 1:(DayIncrementor-1))
        {
        MTMPCT[index-DayIncrementor+j]<<-PercentageChange(MTMT[index-DayIncrementor+j-1],MTMT[index-DayIncrementor+j])
    }
    
    
}
```

```r
PurchasedDataNew <- function(Present_Row,DayIncrementor,BuyStrategy)
{
    
    DateToBuy=TenStockData$Date[Present_Row]
    
    if(DateToBuy=="2018-01-02")
        {
        
        MTMPCT[Present_Row]<<-0
        
        buy(DateToBuy)  
        MTMT[Present_Row]<<-MTM
    }
     else{
         #
         UpdateDividend(Present_Row)
         
         #Find5MostDroppedAdjacentClose and Updates the catalog
         UpdateInitialCatalog(Present_Row,BuyStrategy,DayIncrementor)
         if(DayIncrementor>1)
            {
             DailyMTMCalculator(Present_Row,DayIncrementor)
             }
         
         SellAllShares(Present_Row)
         #empties the portfolio
         PurchasedData<<-PurchasedData[0,]
         #for next purchase
         CashForEachShare<<-InitialCash/5;
         #buy again
         buy(DateToBuy) 
         
         MTMT[Present_Row]<<-MTM
         MTMPCT[Present_Row]<<-PercentageChange(MTMT[Present_Row-1],MTMT[Present_Row])
     }
    
    Present_Row<<-Present_Row+DayIncrementor
}
```

```r
#Calculating HighTechIndex
HighTechIndexCalculator<- function()
    {
    HighTechIndexT=c()
    share = c("IBM", "MSFT", "GOOG", "AAPL", "AMZN", "FB", "NFLX", "TSLA", "ORCL", "SAP");
    TotalShare=length(share)
    for(j in 1:length(TenStockData$Date))
        {
        SumOfTodayShareValue=0
        for(i in 1:length(share))
            {
                curr_column = paste0( share[i], ".Close")
                Current_Share_Price= TenStockData[j, curr_column]
                SumOfTodayShareValue=SumOfTodayShareValue+Current_Share_Price
            }
        DateT[j]<<- TenStockData$Date[j]
        HighTechIndexT[j]=SumOfTodayShareValue/TotalShare
        if(j==1)
            {
            HighTechIndexPCT[j]<<-0
        }
        else
            {
            HighTechIndexPCT[j]<<- PercentageChange(HighTechIndexT[j-1],HighTechIndexT[j])
            }
        }
}
```

```r
#Buying_low strategy 
reset()
DayIncrementor=5
repeat
{
    
    PurchasedDataNew(Present_Row,DayIncrementor,"buying_low")
    print(PurchasedData)
    print(paste("Cash= ",InitialCash))
    print(paste("StockValue= ",TotalStockValue))
    print(paste("MTM= ",MTM))
    
    print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    if(Present_Row>length(TenStockData$Date))
        {
        break
    }
}
print(paste0("The MTM at the end of year with ","buying_low strategy = ",MTM))
```
